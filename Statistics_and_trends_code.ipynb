```python
# Function to create categorical plots
def categorical_plots(data):
    # Plot 1: Bar chart for Outcome distribution
    plt.figure(figsize=(8, 6))
    sns.countplot(data=data, x='Outcome')
    plt.title('Distribution of Diabetes Outcome')
    plt.xlabel('Diabetes Outcome (0 = No, 1 = Yes)')
    plt.ylabel('Count')
    plt.show()

    # Plot 2: Bar chart for Pregnancies distribution in diabetic and non-diabetic cases
    plt.figure(figsize=(10, 6))
    sns.countplot(data=data, x='Pregnancies', hue='Outcome', palette='viridis')
    plt.title('Pregnancies by Diabetes Outcome')
    plt.xlabel('Number of Pregnancies')
    plt.ylabel('Count')
    plt.legend(title='Outcome', labels=['Non-Diabetic', 'Diabetic'])
    plt.show()
```

```python
# Function to create relational plots
def relational_plots(data):
    # Plot 1: Scatter Plot - Glucose vs Insulin
    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x='Glucose', y='Insulin', hue='Outcome', palette='coolwarm', edgecolor="w", s=70)
    plt.title('Glucose vs. Insulin by Diabetes Outcome')
    plt.xlabel('Glucose Level')
    plt.ylabel('Insulin Level')
    plt.legend(title='Diabetes Outcome')
    plt.show()

    # Plot 2: Line Plot - Age vs. BMI (Mean BMI per Age Group)
    mean_bmi_per_age = data.groupby('Age')['BMI'].mean().reset_index()
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=mean_bmi_per_age, x='Age', y='BMI', color='b', marker="o")
    plt.title('Average BMI Across Different Ages')
    plt.xlabel('Age')
    plt.ylabel('Average BMI')
    plt.show()
```

```python
# Function to create statistical plots
def statistical_plots(data, correlation_matrix):
    # Plot 1: Heatmap of Correlation Matrix
    plt.figure(figsize=(12, 10))
    sns.heatmap(correlation_matrix, annot=True, cmap='YlGnBu', fmt=".2f")
    plt.title('Correlation Matrix Heatmap')
    plt.show()

    # Plot 2: Boxplot of BMI by Outcome
    plt.figure(figsize=(8, 6))
    sns.boxplot(data=data, x='Outcome', y='BMI', palette='muted', hue='Outcome', dodge=False)
    plt.title('BMI Distribution by Diabetes Outcome')
    plt.xlabel('Diabetes Outcome (0 = No, 1 = Yes)')
    plt.ylabel('BMI')
    plt.legend([],[], frameon=False)
    plt.show()
```

```python
categorical_plots(df)
```

```python
relational_plots(df)
```

```python
statistical_plots(df, correlation_matrix)
```
